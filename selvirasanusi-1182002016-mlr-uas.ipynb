```r
mlr$Product_holding <- as.factor(mlr$Product_holding)
```

```r
set.seed(733)
```

```r
ind <- sample(2, nrow(mlr), replace = TRUE, prob = c(0.7, 0.3))
```

```r
training <- mlr[ind==1,]
testing <- mlr[ind==2,]
```

```r
library(nnet)
```

```r
training$Product_holding <- relevel(training$Product_holding, ref="1")
```

```r
mlr <- multinom(Product_holding~., data = training)
```

```r
summary(mlr)
```

```r
z <- summary(mlr)$coefficients/summary(mlr)$standard.errors
```

```r
p <- (1 - pnorm(abs(z), 0, 1)) * 2
```

```r
p
```

```r
p <- predict(mlr, training)
```
